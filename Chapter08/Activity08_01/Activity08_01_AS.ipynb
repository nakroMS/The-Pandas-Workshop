```python
#2 Import the pandas, numpy and matplotlib package:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#3 Load the CSV file as a DataFrame:
# file_url = 'https://raw.githubusercontent.com/PacktWorkshops/The-Pandas-Workshop/master/Chapter07/Data/Ames%20Housing.csv'
data_frame = pd.read_csv('Ames_Housing.csv')
```

```python
#4 Display the first 10 rows of the DataFrame:
# data_frame.head(10)
data_frame.head(10)
```

```python
#5 Plot an histogram chart for sale_price:
# data_frame.sale_price.plot(kind = 'hist');
data_frame.sale_price.plot(kind = 'hist');
```

```python
#6 Plot an histogram chart for lot_area:
# data_frame.lot_area.plot(kind = 'hist');
data_frame.lot_area.plot(kind = 'hist');
```

```python
#7 Plot a scatter plot chart for sale_price and lot_area:
# data_frame.plot(kind='scatter', y = 'sale_price', x = 'lot_area');
data_frame.plot(kind = 'scatter', y = 'sale_price', x = 'lot_area');
```

```python
#8 Plot a boxplot chart for overall_condition on sale_price:
# data_frame.boxplot(by='overall_condition', column='sale_price');
data_frame.boxplot(by = 'overall_condition', column = 'sale_price');
```

```python
#9 Plot a scatter chart for overall_condition and sale_price:
# data_frame.plot(kind='scatter', x = 'overall_condition', y = 'sale_price');
data_frame.plot(kind = 'scatter', x = 'overall_condition', y = 'sale_price'); 
```

```python
#10 Plot a boxplot chart for overall_quality on sale_price:
# data_frame.boxplot(by='overall_quality', column='sale_price');
data_frame.boxplot(by = 'overall_quality', column = 'sale_price');
```

```python
#11 Plot a scatter chart for overall_quality on sale_price:
data_frame.plot(kind='scatter', x = 'overall_quality', y = 'sale_price');
```

```python
#12 Plot a boxplot chart for overall_quality on year_built:
data_frame.boxplot(by='overall_quality', column='year_built');
```

```python
#13 Plot an horizontal bar chart for neighborhood and sale_price by using the pivot function:
# pd.pivot_table(data_frame,index=['neighborhood']).sale_price.plot(kind='barh');
pd.pivot_table(data_frame, index = ['neighborhood']).sale_price.plot(kind = 'barh');
```

```python
#14 Plot an horizontal bar chart for neighborhood and overall_quality by using the pivot function:
pd.pivot_table(data_frame,index=['neighborhood']).overall_quality.plot(kind='barh');
```

```python
#15 Plot a line chart for year_built and sale_price by using the pivot function:
# pd.pivot_table(data_frame,index=['year_built']).sale_price.plot();
pd.pivot_table(data_frame, index = ['year_built']).sale_price.plot();
```

```python
#16 Plot a line chart for year_sold and sale_price by using the pivot function:
pd.pivot_table(data_frame,index=['year_sold']).sale_price.plot();
```

```python
#17 Plot a scatter plot chart for sale_price, overall_condition and overall_quality by using the c and colormap='Blues' parameter:
# data_frame.plot(kind='scatter', y = 'sale_price', x = 'overall_condition', c = 'overall_quality',colormap='Blues');
data_frame.plot(kind = 'scatter', y = 'sale_price', x = 'overall_condition', c = 'overall_quality', colormap = 'Blues');
```
